```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# plotting style
%config InlineBackend.figure_format = 'retina'
plt.style.use('paper.mplstyle')
twocol = 7.1014
onecol = 3.35
```

```python
def log_norm(x,mode,sigma,A):
    mu = np.log(mode) + sigma**2
    f1 = lambda x: 1e5/(x*sigma*np.sqrt(2*np.pi))*np.exp(-(np.log(x)-mu)**2/(2*sigma**2))
    f = lambda x: A/f1(mode)*f1(x)
    return f(x)
```

```python
x = np.linspace(0.1,20000,1000)

"""
params = [[1000,   1,25],
          [1000,0.95,20],
          [5000, 0.6, 4]]
"""

N = 5
modes = np.linspace(1000,5000,N)
sigmas = np.geomspace(0.8,0.5,N)
As = np.linspace(25,4,N)

for i in range(N):
    plt.plot(x,log_norm(x,modes[i],sigmas[i],As[i]))

plt.show()
```
