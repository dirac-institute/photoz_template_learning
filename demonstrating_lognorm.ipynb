```python
def log_norm(x,mode,sigma,A):
    mu = np.log(mode) + sigma**2
    f1 = lambda x: 1e5/(x*sigma*np.sqrt(2*np.pi))*np.exp(-(np.log(x)-mu)**2/(2*sigma**2))
    f = lambda x: A/f1(mode)*f1(x)
    return f(x)
```

```python
x = np.linspace(0.1,20000,1000)

params = [[1000,   1,25],
          [1000,0.95,20],
          [5000, 0.6, 4]]

for p in params:
    plt.plot(x,log_norm(x,p[0],p[1],p[2]))
#plt.plot(x,log_norm(x,1000,1,25))
#plt.plot(x,log_norm(x,1000,0.95,20))
#plt.plot(x,log_norm(x,5000,0.5,5))

plt.show()
```
